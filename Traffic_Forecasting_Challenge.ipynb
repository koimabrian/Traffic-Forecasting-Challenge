```python
from google.colab import drive
drive.mount('/content/drive')

import pandas as pd

# Define the path
path = '/content/drive/MyDrive/Colab Notebooks/zindi/Spatio/data/'

# Load the datasets
traffic_DLThpVol = pd.read_csv(path + 'traffic_DLThpVol.csv')
traffic_DLThpTime = pd.read_csv(path + 'traffic_DLThpTime.csv')
traffic_DLPRB = pd.read_csv(path + 'traffic_DLPRB.csv')
traffic_MR_number = pd.read_csv(path + 'traffic_MR_number.csv')
SampleSubmission = pd.read_csv(path + 'SampleSubmission.csv')

# Print columns for each dataset
print("Columns in traffic_DLThpVol.csv:", traffic_DLThpVol.columns.tolist())
print("Columns in traffic_DLThpTime.csv:", traffic_DLThpTime.columns.tolist())
print("Columns in traffic_DLPRB.csv:", traffic_DLPRB.columns.tolist())
print("Columns in traffic_MR_number.csv:", traffic_MR_number.columns.tolist())
print("Columns in SampleSubmission.csv:", SampleSubmission.columns.tolist())
```

```python
# Function to summarize a dataset
def summarize_data(df, name):
    print(f"Summary for {name}:")
    print(f"Number of rows: {df.shape[0]}")
    print(f"Number of columns: {df.shape[1]}")
    print(f"Column names: {df.columns.tolist()}")
    print(f"Data types:\n{df.dtypes}")
    print(f"Missing values:\n{df.isnull().sum()}")
    print(f"First few rows:\n{df.head()}")
    print("-" * 50)

# Summarize each dataset
summarize_data(traffic_DLThpVol, 'traffic_DLThpVol.csv')
summarize_data(traffic_DLThpTime, 'traffic_DLThpTime.csv')
summarize_data(traffic_DLPRB, 'traffic_DLPRB.csv')
summarize_data(traffic_MR_number, 'traffic_MR_number.csv')
```
